```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Carregamento dos dados de Teste

```python
df = pd.read_csv("../models/test_data.csv")
df.head()
```

### Tabela Verdade

```python
# Gerar crosstab dinamicamente
ct = pd.crosstab(df.label_verdadeiro, df.label_predito, margins=True)

# Separar a matriz principal (sem totais) e os totais
ct_main = ct.iloc[:-1, :-1]

# Plot
plt.figure(figsize=(8, 6))

sns.heatmap(
    ct_main,
    annot=True,
    fmt="d",
    cmap="Blues",
    cbar=True,
    linewidths=0.5,
    linecolor='gray'
)

plt.title("Matriz de Confusão (Predito vs Verdadeiro)")
plt.xlabel("Predito")
plt.ylabel("Verdadeiro")

plt.tight_layout()
plt.show()
```

```python
errors = df[df.label_verdadeiro != df.label_predito]
errors.head(10)
```
